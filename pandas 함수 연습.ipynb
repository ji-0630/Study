### Q57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑<span style="color:red"> Re </span>

* Unknown : N 
* Less than 40K : a
* 40K - 60K : b
* 60K - 80K : c
* 80K - 120K : d
* 120K + : e

````python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}


card['newIncome'] = card['Income_Category'].map(lambda x: dic[x])
card['newIncome']
````

### Q58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑<span style="color:red"> Re </span>

* Unknown : N 
* Less than 40K : a
* 40K - 60K : b
* 60K - 80K : c
* 80K - 120K : d
* 120K + : e

````python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'

card['newIncome']  = card.Income_Category.apply(changeCategory)
card['newIncome']
````

### Q62. Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의. newState의 각 값들의 빈도수를 출력<span style="color:red"> Re </span>

### Q67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의

### Q76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다<span style="color:red"> Re </span>

````python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
dust = pd.read_csv(DataUrl)
dust['(년-월-일:시)'].head()
````

````python
def change_date(x):
    import datetime
    
    hour = x.split(':')[1]
    date = x.split(":")[0]
    
    if hour =='24':
        hour ='00:00:00'
        
        FinalDate = pd.to_datetime(date +" "+hour) +datetime.timedelta(days=1)
        
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)
    
    return FinalDate

dust['(년-월-일:시)'] = dust['(년-월-일:시)'].apply(change_date)
pd.to_datetime(dust['(년-월-일:시)'])
dust.head()
````
