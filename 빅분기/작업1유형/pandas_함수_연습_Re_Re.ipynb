### Q57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑

```python
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
 ```

````python
dic = {'Unknown'        : 'N',
       'Less than $40K' : 'a',
       '$40K - $60K'    : 'b',
       '$60K - $80K'    : 'c',
       '$80K - $120K'   : 'd',
       '$120K +'        : 'e' }

card['newIncome'] = card['Income_Category'].map(lambda x: dic[x])
card.head()
````

### Q58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑

```python
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
 ```

````python
def dic(x):
  if x == 'Unknown':
    return 'N'
  elif x ==  'Less than $40K':
    return 'a'
  elif x==     '$40K - $60K' :
    return 'b'
  elif x ==   '$60K - $80K' :
    return 'c'
  elif x ==  '$80K - $120K':
    return 'd'
  else:
    return 'e'

card['newIncome'] = card['Income_Category'].apply(dic)
card['newIncome']
````

### Q60. Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력

````python
def edu(x):
  if x in 'Graduate':
    return 1
  else:
    return 0

card['newEduLevel'] = card['Education_Level'].apply(edu)
card['newEduLevel'].value_counts()
````

### Q76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다<span style="color:red"> Re </span>

````python
def ch_time(x):
  import datetime as dt

  date = x.split(":")[0]
  hour = x.split(":")[1]

  if hour == '24':
    fin = date + " 00:00:00"
    fin = pd.to_datetime(fin) + dt.timedelta(days=1)
  else:
    fin = date + " " + hour + ":00:00"
    fin = pd.to_datetime(fin)

  return fin

seoul['(년-월-일:시)'] = seoul['(년-월-일:시)'].apply(ch_time)

seoul['(년-월-일:시)'] 
````
